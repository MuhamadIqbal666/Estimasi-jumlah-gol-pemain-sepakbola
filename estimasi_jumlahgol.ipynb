#Jumlah gol sesuai nama club

```python
Club = df.groupby('Club').count()[['Goals']].sort_values(by='Goals',ascending=True).reset_index()
Club = Club.rename(columns={'Goals':'Goals'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=Club['Club'], y=Club['Goals'], color='royalblue')
plt.xticks(rotation=60)
```

```python
Goals= df.groupby('Goals').count()[['Matches_Played']].sort_values (by='Matches_Played').reset_index()
Goals= Goals.rename(columns={'Matches_Played':'Matches_Played'})
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x= Goals['Goals'], y= Goals['Matches_Played'], color='royalblue')
plt.xticks(rotation=60)
```

# Seleksi Fitur

```python
feature = ['Mins',	'Matches_Played',	'Shots',	'OnTarget',	'Year']
x = df[feature]
y = df['Goals']
x.shape, y.shape
```

# Split Data Training dan Data Testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# Membuat Model Regresi Linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regesi linier =', score)
```

#Membuat Inputan Model Regresi Linier

```python
#Mins= 2071, Matches_Played=22,  Shots=37, OnTarget=26,  Year=2019
input_data = np.array([[2071,22,37,26,2019]])

prediction = lr.predict(input_data)
print('Estimasi Gol yang dihasilkan permusim :', prediction)
```

# Save Model

```python
import pickle
filename = 'estimasi_jumlahgol.sav'
pickle.dump(lr,open(filename,'wb'))
```

# Evaluasi

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Memasukkan fitur dan target
feature = ['Mins', 'Matches_Played', 'Shots', 'OnTarget', 'Year']
x = df[feature]
y = df['Goals']

# Membuat model dan melatihnya
model = LinearRegression()
model.fit(x, y)

# Melakukan prediksi menggunakan model
y_pred = model.predict(x)
r_squared = r2_score(y, y_pred)

# Menampilkan R-squared
print("Nilai akurasi R-squared (R^2):", r_squared)
```
